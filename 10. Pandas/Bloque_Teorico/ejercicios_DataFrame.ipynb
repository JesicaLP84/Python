```python
import pandas as pd
import numpy as np
brand = pd.Series(['Camp', 'Camp', 'Petzl', 'Petzl', 'Edelrid', 'Edelrid', 'Edelrid', 'Black Diamond', 'Black Diamond','Mammut'])
models = pd.Series(['Energy', 'Jasper', 'Simba', 'Adjama New', 'Moe', 'Orion', 'Leaf','Xenes', 'Chaos', 'Ophir'])
prices = pd.Series([39.90, 56.00, 45.00, 75.00, 49.90, 99.90, 65.00, 119.90, 99.90, 55.00])
```

```python
#1.Usando las listas brands, models y prices de los ejercios anterior. Crear un DataFrame llamado que almacene toda esta información en tres columnas (brand, model y price) y los índices los genere automáticamente.
dataFrame= pd.DataFrame({ 'brand' : brand,
    'models' : models,
    'prices' : prices
}, columns=['brand', 'models','prices'])
dataFrame
```

```python
#2.Mostrar la columna “Brand”.
dataFrame['brand']
#otro metodo es : 
# dataframe.brand
```

```python
#3.Muestra la quinta fila.
dataFrame.loc[5]
#otra forma es:
# dataFrame.iloc[5]
```

```python
#4.Modifica el precio del modelo “Energy” a 9999.
  
dataFrame['prices'] = dataFrame['prices'].replace([39.9],[9999])
dataFrame
```

```python
#5.Crea una nueva columna “Discount” con el valor 50 para cada fila.
dataFrame['Discount'] = 50
dataFrame
```

```python
#6.Usando el DataFrame del Ejercicio 5, crea un nuevo DataFrame que añada las columnas “Sales” y “Total”.
dataFrame_new=dataFrame.assign(Sales=0,Total=0)
dataFrame_new
```
