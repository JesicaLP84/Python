```python
import numpy as np
import pandas as pd
```

```python
#1.- Obtener información del archivo trade_details:
read_trade_details = pd.read_csv('trade_details.csv', sep=';') # indico el separador porque viene con ; y entonces no es del todo legible, lo cambio a , para que se muestr bien columnas y valores
read_trade_details
```

```python
#2.- Obtener información del archivo trade_details_snapshot:
read_trade_details_snapshot = pd.read_csv('trade_details_snapshot.csv', sep=';')
read_trade_details_snapshot
```

```python
#3.- Ordenar ambos archivos la columna de maturity de forma descendente para saber cual es el dato más actual
#*long leg*, es decir, la que tiene una mayor fecha de vencimiento (fecha más actual de la columna maturity).
print(read_trade_details.sort_values('maturity', ascending=False))
read_trade_details_snapshot.sort_values('maturity', ascending=False)
```

```python
#4.- Unir ambos Dataframe creando un DF donde se registraran todos los datos finales

#El método combina dos DataFrame y utiliza un valor del DataFrame si el primero tiene un valor NULL 
trade_fusion = pd.DataFrame(read_trade_details.combine_first(read_trade_details_snapshot))
trade_fusion

#--------------------ESTE METODO LO UTILIZE INICIALMENTE PARA HACER EL MERGE DE AMBOS DATAFRAME--------------
# trade_combi = pd.merge(read_trade_details,read_trade_details_snapshot, how='outer').sort_values('maturity')
# trade_combi 
```

```python
#5.- Filtro snapshot por la columna de mFamily = CURR para comprobar de los FXSwaps cuales estan con distinto trade_number y un mismo contract_number para obtener solo la *long leg*

filter_dfFusuion = trade_fusion.loc[:,'mfamily'] == 'CURR'
filter_FXSwaps = trade_fusion.loc[filter_dfFusuion]
filter_FXSwaps

# ----_ESTE FILTRO ERA UTILIZADO CON DATAFRAME DE LA COPIA DE SEGURIDAD_---
# filtro = read_trade_details_snapshot.loc[:,'mfamily'] == 'CURR'
# filtro_oden = read_trade_details_snapshot.loc[filtro]
# filtro_oden
```

```python
#6.-Eliminar del DataFrame Fusionado las filas que contengan la data short y asi solo dejar reflejadas las long leg

deleteFXSwaps_1 = trade_fusion[ (trade_fusion['origin_contract_number'] == 18622136) & (trade_fusion['maturity']== '2020-04-29')].index
deleteFXSwaps_2 = trade_fusion[ (trade_fusion['origin_contract_number'] == 18724280) & (trade_fusion['maturity']== '2020-11-05')].index
deleteFXSwaps_3 = trade_fusion[ (trade_fusion['origin_contract_number'] == 19883451) & (trade_fusion['maturity']== '2020-01-19')].index
deleteFXSwaps_4 = trade_fusion[ (trade_fusion['origin_contract_number'] == 21622649) & (trade_fusion['maturity']== '2021-05-04')].index
deleteFXSwaps_5 = trade_fusion[ (trade_fusion['origin_contract_number'] == 19622128) & (trade_fusion['maturity']== '2020-04-29')].index

trade_fusion.drop(deleteFXSwaps_1, inplace=True)
trade_fusion.drop(deleteFXSwaps_2, inplace=True)
trade_fusion.drop(deleteFXSwaps_3, inplace=True)
trade_fusion.drop(deleteFXSwaps_4, inplace=True)
trade_fusion.drop(deleteFXSwaps_5, inplace=True)

trade_fusion
```
